Bovine tuberculosis (bTB) is a disease that threatens livestock and thereby has potentially disasterous effects on farming and (to a leser extenet) economies and public health. Great Britain in particular has devoted substantial resources to research and control bTB. People have long posited that Eurasian badgers (*Meles meles*) contribute in very serious ways to infection in cattle--the natural reaction to which, of course, is to cull the badgers. The erradication of some 20,000 badgers in Great Britain between the years 1975 and 1997 [cite], nevertheless, has had questionable results. In 2005, D.R. Cox *et. al* proposed a simple model of the prevelance and incidence of bTB in cattle. This model was based on relatively basic linear dynamics concepts (among math from a few other disciplines). More recent research on the topic (both by Cox and his coauthors as well as other academics), however, has considerably underminded the aforementioned model's elegant and optimistic results. Modeling this phenomenon, it appears, has less to do with the number of living badgers than other tremendously complex factors like their social behavior and spacial distribution as well as the distribution of cattle.

Cox *et. al*'s model relies on a system of non-homogeneous linear differential equations with a lot of simplifying assumptions. Consider the system given by:

$$
    \dot{\vec{X}}(t) = (\vec{A} - \vec{R})\vec{X}(t) + \vec{S}
$$

where the rows of $\vec{X}(t)$ are $C(t)$ (the number of infected cattle per $km^2$) and $B(t)$ (the number of infected badgers per $km^2$), $\vec{A}$ is $a_{ij}$, the rate (per year, typically) that an animal of species $j$ infects an animal of species $j$, $\vec{R}$ is $\text{diag}{(r_1, r_2)}$, where $r_i$ is the removal rate for animals of species $i$ (constant for the period in question), and $\vec{S}$ is the column vector $s_i$ (the restocking of species $i$). Constant restocking and removal (both by routine slaughter and as a result of discovery of bTB) are simplifications, but altogether sensible ones. The rate of removal is contributed to by two things: the amount of routtine slaughter and the amount of animals slaughtered to remove reactor (infected) cattle. The rate of removal of reactor cattle will remain constant as long as testing frequency and test sensitivity remain constant (this model stipulates that they do). The restocking rate is assumed to be constant (I suppose simply because it makes sense for farmers not to buy more or fewer cattle between years).

This equation yeilds the following solution (appendix):

$$
    C(t) = C_0 e^{\lambda - r_1} + \int_{0}^{t} e^{(\lambda - r_1)t}[a_{1 2}B(t - x) + s_1(t - x)] dx,
$$

where $\lambda$ is an eigenvalue of $\vec{A}$. This, mind you, makes no assumptions about a culling. In a situation where the system starts at equilibrium values ($\vec{X_0} = \vec{X}^* = (C^*, B^*)$) and then, at $t = 0$, there is a culling that instantly diminishes badger populaton to $(1 - k)B^*$ (and therafter returns to its initial level as $B(t) = B^*(1 - k e^{-v t})$), the general solution is:

$$
    C(t) = C^* + a_{12}B^*k(\lambda - r_1 + v)^{-1}(e^{-vt} - e^{(\lambda-r_1})t).
$$

Numerical solutions are then made dimensionless with $r_1t = \tau$, $I_b = a_{12}B^*/s_1$, $v = \nu/r_1$, and $l = \lambda/r_1$ and dividing through by $C^* = \frac{s_1 + a_{12}B^*}{r_1 - a_{11}}$ (appendix):

$$
    C(t)/C^* = 1 - \frac{k I_B (1 - l)}{(1 + I_B)(1 - l - v)}(e^{-v t} - e^{(1 - l)t})
$$

This equation is, indeed, a standardized prevalence of tBT in cattle.

(fig)

More importantly, arguably, is the value of the latent incidence value. To be sure, this is, in fact, the quantity that the British government intends to lower by giving out badger culling lisences. This quantity is given by

$$
    \dot{C}(t) + r_1 = r_1 C^* - a_{12} B^*(r_1 - k \nu)(r_1 - \lambda - \nu)^{-1}e^{-\nu t} + a_{12} B^*(r_1) - k r_1 + k \lambda)(r_1 - \lambda -\nu)^{-1}e^{(\lambda - r_1)t}
$$

(appendix)

Dividing this equation by the equilibrium rate $r_1 C^*$ and then non-dimensionalizing with the values above gives

$$
    1 - \frac{I_B (1 - k v)(1 - l)}{(1 + I_B)(1 - l - v)} e^{-v t} + \frac{I_B (1 - k + k l)(1 - l)}{(1 + I_B)(1 - l - v)} e^{(1 - l)t}
$$

```python
def plot_curve(curve, ts, *args, **kwargs):
    plt.figure(figsize=FIGSIZE)
    p = plt.plot(ts, curve)
    for k, v in kwargs.items():
        if (k == 'label'): plt.legend(p, [v])
        else: plt.__getattribute__(k)(v)
    plt.legend(loc=4)
```

```python
memp = {}

def prevalence(t, v, k, Ib, l):
    """Uses model to return prevalence as function of time.

    :param v: ratio of badger growth rate to removal rate of cattle
    :param k: percent of badgers killed
    :param Ib: ratio of the number of cattle infected by badgers per year to
    the number of infected cattle brought in from outside of the system per year.
    :param l: ratio of eigenvalue of A to removal rate of cattle
    """
    
    key = (t[0],t[-1],t.size,v,k,Ib,l)
    if memp.has_key(key): return memp.get(key)
    num = k * Ib * (1. - l)
    denom = (1 + Ib)*(1. - l - v)
    val = 1. - (num/denom) * (exp(-v * t) - exp((l-1.) * t))
    memp[key] = val
    return val
```

```python
T = linspace(0, 10, 50)
FIGSIZE = (11, 5)
```

```python
def plot_prevalence(v=0.17, k=0.5, Ib=2.0, l=0.2, ymin=0.6):
    plot_curve(prevalence(T, v, k, Ib, l), T,
        title='standardized bTB prevalence in cattle',
        label=r'$k={},\,I_B={:.2f},\,l={}$'.format(k, Ib, l),
        xlabel=r'$\tau$',
        ylim=(ymin, 1.05))
```

```python
interact(plot_prevalence,
         v=(0.0, 0.3, 0.03),
         k=(0.2, .9, 0.07),
         Ib=(0.00001, 2.0, 0.15),
         l=(0.05, 0.5,0.045),
         ymin=(0.35, 0.9));
```

```python
memi = {}

def incidence(t, v, k, Ib, l):
    """Uses model to return incidence as function of time.

    :param v: ratio of badger growth rate to removal rate of cattle
    :param k: percent of badgers killed
    :param Ib: ratio of the number of cattle infected by badgers per year to
    the number of infected cattle brought in from outside of the system per year.
    :param l: ratio of eigenvalue of A to removal rate of cattle
    """
    key = (t[0], t[-1], t.size, v, k, Ib, l)
    if memi.has_key(key): return memi.get(key)
    num1 = Ib * (1 - k*v) * (1 - l)
    num2 = Ib * (1 - k + k * l) * (1 - l)
    denom = (1 + Ib)*(1. - l - v)
    val = 1. - num1/denom * exp(-v * t) + num2/denom * exp((l-1.) * t)
    memi[key] = val
    return val
```

```python
def plot_incidence(v=0.17, k=0.5, Ib=2.0, l=0.2, ymin=0.4):
    plot_curve(incidence(T, v, k, Ib, l), T,
               title='standardized bTB incidence in cattle',
               xlabel=r'$\tau$',
               ylabel=r'$C(\tau)/C^*$',
               ylim=(ymin, 1.05));
```

```python
interact(plot_incidence,
         v=(0.0, 0.3, 0.03),
         k=(0.2, .9, 0.07),
         Ib=(0.0, 2.0, 0.15),
         l=(0.05, 0.5,0.045),
         ymin=(0.35, 0.9));
```
